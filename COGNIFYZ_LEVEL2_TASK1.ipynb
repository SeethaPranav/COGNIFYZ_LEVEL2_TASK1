## Analyze the distribution of aggregate ratings and find the most common rating range

### 1. Explore unique values and basic stats of 'Aggregate rating'

```python
print(data['Aggregate rating'].describe())
print("\nUnique ratings:", sorted(data['Aggregate rating'].unique()))
```

### 2. Visualize distribution with a histogram

```python
plt.figure(figsize=(8,5))
sns.histplot(data['Aggregate rating'], bins=20, kde=True, color='skyblue', edgecolor='black')

plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Number of Restaurants')
plt.xlim(0, 5)
plt.grid(True)
plt.show()
```

### 3. Find the most common rating

```python
most_common_rating = data['Aggregate rating'].mode()[0]
most_common_count = data['Aggregate rating'].value_counts().iloc[0]

print(f"Most common rating: {most_common_rating} (appears {most_common_count} times)")
```

## Calculate the average number of votes received by restaurants

```python
average_votes = data['Votes'].mean().round(2)

print(f"Average number of votes per restaurant: {average_votes}")
```
